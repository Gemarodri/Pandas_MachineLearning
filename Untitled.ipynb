```python
#imports
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#Let's first check gender
titanic_df.groupby(['Sex']).size()
```

```python
#Now we want to separate the genders by classes
titanic_df.groupby(['Sex','Pclass']).size()
sns.factorplot('Sex',data=titanic_df, hue='Pclass',kind='count')
sns.factorplot('Pclass',data=titanic_df, hue='Sex',kind='count')
```

```python
#Create a function were we can see the male,famale and childrens in the boat.
def male_female_child(passenger):
    age,sex = passenger
    
    if age < 16:
         return 'child'
    else:
        return sex
```

```python
#We are goung to define a new colum call person to define male,famele or child
titanic_df['person'] = titanic_df[['Age','Sex']].apply(male_female_child, axis=1)

#plot 
sns.factorplot('Pclass',data=titanic_df, hue='person',kind='count')
```

```python
#We are going to create a distribution of the ages to get a more precise picture of who the passengers really were
#We can create a quick histogram
titanic_df['Age'].hist(bins=70)
```

```python
#Which was the mean Age of the passengers
titanic_df['Age'].mean()
```

```python
#Let's go ahead and get a qucik overall comparison of male, female, child
titanic_df['person'].value_counts()

#The same way
titanic_df.groupby('person').size()
```

# 2. What deck (piso/cubierta) were the passengers on and how does that relate to their class ?

```python
titanic_df.head()
```

```python
#Cabin column has the information about the deck, but it has a lot of NaN which we should remove
#We are going to do that creating a new onject called deck,
deck=titanic_df['Cabin'].dropna()
deck.head()
```

```python
# The levels are A,B,C,D,E,F,G, so we just need the first letter to know wich cabin the passenger was
# Let's do that by using a loop

levels = []
for level in deck:
    levels.append(level[0])

cabin_df = DataFrame(levels)
cabin_df.columns=['Cabin']
sns.factorplot('Cabin', data=cabin_df,palette='winter_d',kind='count')

#The last group the T does not have sense so we are gonna to drop it
cabin_df = cabin_df[cabin_df.Cabin !='T']
sns.factorplot('Cabin', data=cabin_df,palette='summer',kind='count')
```

# 4. Who was alone and who was with family?

```python
# We can start by adding a new column to define what alone means.
# For us, alone means that the passenger has 0 in the SibSP and Parch columns
titanic_df['Alone'] = titanic_df.SibSp + titanic_df.Parch
#Any value different from 0 in the Alone colum means that the passenger has family abroad
```

```python
titanic_df['Alone'].loc[titanic_df['Alone']>0] = 'With Family'
titanic_df['Alone'].loc[titanic_df['Alone']==0] = 'Alone'
```

```python
#visualization
sns.factorplot('Alone',data=titanic_df,palette='Blues', kind='count')
```

# 5. What factors helped soemeone survive the sinking?

```python
#We can create a new colum and we can do that through mapping
titanic_df['Survivor'] = titanic_df.Survived.map({0:'no', 1:'yes'})
```

```python
titanic_df.head()
sns.factorplot('Survivor',data=titanic_df,palette='Set1', kind='count')
```

```python
#Let's see if the class was a factor to survive or no
#is the class a factor?
sns.factorplot('Pclass','Survived',data=titanic_df)
```

```python
#Tis the Sex a factor?
sns.factorplot('Pclass','Survived',hue='person',data=titanic_df)
```

We can observe taht even regardless of the class, the result of being a male in any on the classes dramatically decreases the probability of survive, But now we would ask if the age is a factor

```python
#Lineal plot age vs survivor
sns.lmplot('Age','Survived', data=titanic_df)
```

The line on the 1.o are all the people that survived and their corresponding age
On the bottom, in the 0.0 line is everyone who did not survive and their corresponding age
There's a general trend if you look at the lineal line, the older teh passanger was the less likely they survive, that's the general trend

```python
# Let's go to use hue to see the effects of class and age
sns.lmplot('Age','Survived', hue='Pclass', data=titanic_df, palette='winter')
```

```python
#Looks like there's still quite a bit mre deadths in the third class and we can use the x bin 
#argument to clean up this figure a little bit more.

generation = [10,20,40,50,80]
sns.lmplot('Age','Survived', hue='Pclass', data=titanic_df, palette='winter', x_bins=generation)
```

You can seee that there is a high standar desviation on the older people in the firsta class which is an interesting find.

```python
sns.lmplot('Age','Survived',hue='Sex', data=titanic_df, palette='winter', x_bins=generation)
```

Interesting! It looks that if you are an older women you had a better chance of surviving versus if you were an older male. But still the very old male has a quite a bit of standard deviation on it, so it might be interesting to get a closer look on that.
